## EDA On Netflix data

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python

df=pd.read_csv('NetflixOriginals.csv',encoding='latin-1')
df.head(5)
```

```python
df.info()
```

```python
# Check Any Missing value
df.isnull().sum()
```

```python

# Check Duplicate
df.duplicated()
```

```python
# Statistics Analysis
df.describe()
```

```python
# Histogram OF IMDB Score
df_IMDB=df['IMDB Score']
df_IMDB.hist()
```

```python
# Count Total Genre
df['Genre'].count()
```

```python
# Plot Frequency
df['Premiere']=pd.to_datetime(df['Premiere'],format='mixed')
df['Premiere_year']=df['Premiere'].dt.year


plt.figure(figsize=(10,6))
year_count=df['Premiere_year'].value_counts().sort_index()
year_count.plot(kind='bar',color='blue',edgecolor='black')
plt.title("Netflix Originals By Premiere Year",fontweight='bold',fontsize=16)
plt.xlabel("Premiere Year",fontweight='bold',fontsize=14)
plt.ylabel("Number of Premiere",fontweight='bold',fontsize=14)

```
