## Plot Important Graphs

```python
# Histogram OF IMDB Score
df_IMDB=df['IMDB Score']
df_IMDB.hist()
```

```python
# Count Total Genre
df['Genre'].count()
```

```python
# Plot Frequency
df['Premiere']=pd.to_datetime(df['Premiere'],format='mixed')
df['Premiere_year']=df['Premiere'].dt.year


plt.figure(figsize=(10,6))
year_count=df['Premiere_year'].value_counts().sort_index()
year_count.plot(kind='bar',color='blue',edgecolor='black')
plt.title("Netflix Originals By Premiere Year",fontweight='bold',fontsize=16)
plt.xlabel("Premiere Year",fontweight='bold',fontsize=14)
plt.ylabel("Number of Premiere",fontweight='bold',fontsize=14)

```

```python
# To Compare Runtime and IMDB Score 1st make Datatype same for both column
# ! Change datatype of Runtime from int to float
df['Runtime']=df['Runtime'].astype(float)
df.dtypes
```

```python
# Compare Runtime and IDMB Score with Scatter Plot

plt.figure(figsize=(8,6))
plt.scatter(df['Runtime'],df['IMDB Score'],color='Red',edgecolors='Yellow')
plt.title("Runtime VS IMDB Score")
plt.xlabel("Runtime (Min)")
plt.ylabel("IMDB Score")
plt.grid(True)
```
