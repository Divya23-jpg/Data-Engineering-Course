Identify IQR

```python

# Numpy To Identify the IQR

q1=np.percentile(required_column['Radius (mean)'],25)
q3=np.percentile(required_column['Radius (mean)'],75)
iqr=q3-q1
iqr
```

```python
# ! Pandas to Identify the IQR
q1=required_column['Radius (mean)'].quantile(0.25)
q3=required_column['Radius (mean)'].quantile(0.75)
iqr=q3-q1
iqr
```

```python
lower_bound=q1-(1.5*iqr)
higher_bound=q3+(1.5*iqr)
print(lower_bound)
print(higher_bound)
```

```python

final_Radius=required_column[(required_column['Radius (mean)']>= lower_bound) & (required_column['Radius (mean)']<=higher_bound)]
final_Radius.describe()
```

## Visual Representation of Outlier

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np

np.random.seed(42)
x=np.random.normal(50,10,100)
y=np.random.normal(0,5,100)


df = pd.DataFrame({'X': x, 'Y': y})
df.iloc[5]=[50,200]
df.iloc[25]=[90,20]

Q1=df.quantile(0.25)
Q3=df.quantile(0.75)
IQR=Q3-Q1

is_inside_bounds=(df>=(Q1-1.5*IQR)) & (df<=(Q3+1.5*IQR))
df['is_outlier']=~is_inside_bounds.all(axis=1)

fig, axes = plt.subplots(1, 2, figsize=(12, 6))
sns.scatterplot(data=df,x='X',y="Y",hue='is_outlier',ax=axes[0])
axes[0].set_title("Scatter Plot with Outlier")

df_no_outlier=df[~df['is_outlier']]
sns.scatterplot(data=df_no_outlier,x='X',y="Y",ax=axes[1])
axes[1].set_title("Scatter Plot Without Outlier")

plt.tight_layout()
plt.show()
```
