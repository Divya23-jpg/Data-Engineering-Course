###Partition BY Range

```
data_frame2=spark.read.option("header",True).csv("/content/NetflixOriginals (1).csv",inferSchema=True)
```

```
data_frame2.repartitionByRange(4,"Premiere")
```

```
print(data_frame2.rdd.glom().collect())
```

```
data_frame2.show(5)
```
