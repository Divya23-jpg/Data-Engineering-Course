## Exersice for Data Frame Transformations

1.Select specific Colums (id,quantities,price)

2.Perform Join Operation with Id as Commom attribute for df1 and df2.

3.Sort the data on the basis of price.

4.Check the Distinct categories.

5.Renaming the Column (Name) change to Product_name

```
%%bash
head -10 /content/stocks.txt
```

```
path_txt="/content/stocks.txt"
df_txt=spark.read.csv("/content/stocks.txt",header=True,inferSchema=True)
df_txt.show(5)
```

###1.Select specific Colums (id,quantities,price)

```
specific_column = df_txt.select("id", "name", "quantity")
specific_column.show(5)
```

###

###2.Perform Join Operation with Id as Commom attribute for df1 and df2.

```
df1 = df_txt.select("id", "name", "quantity")
df2 = df_txt.select("id","category", "price")
joined_data=df1.join(df2,on='id')
joined_data.show(5)
```

###3.Sort the data on the basis of price

```
sorted_data=df_txt.orderBy('price',)
sorted_data.show(5)
```

###4.Check the Distinct categories.

```
distinct_value=df_txt.select('category').distinct()
distinct_value.show()
```

###5.Renaming the Column (Name) change to Product_name

```
renamed_data=df_txt.withColumnRenamed('name','product_name')
renamed_data.show(5)
```
