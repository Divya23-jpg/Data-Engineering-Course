###PySpark Tutorial

```
import pyspark
from pyspark.sql import SparkSession
import pandas as pd
import numpy as np
```

```
spark=SparkSession.builder.appName('demo-app').getOrCreate()
```

```
csv_file_path="/content/products.csv"
df=spark.read.csv(csv_file_path,header=True)
```

```
df.show(5)
```

```
df.printSchema();
```

```
from pyspark.sql.types import StructType , StructField,StringType,DoubleType,IntegerType
```

```
schema=StructType(
    [
        StructField(name='id',dataType=IntegerType(),nullable=True),
        StructField(name='name',dataType=StringType(),nullable=True),
        StructField(name='category',dataType=StringType(),nullable=True),
        StructField(name='quantity',dataType=IntegerType(),nullable=True),
        StructField(name='price',dataType=DoubleType(),nullable=True)
    ]
)
```

```
csv_file_path="/content/products.csv"
```

```
df_with_schema=spark.read.csv(csv_file_path,header=True,schema=schema)
```

```
df_with_schema.printSchema()
```

```
%%bash
head -10 /content/products_multiline.json
```

```
df_json_data_multiline = (
    spark.read
         .option("multiline", "true")   # <-- correct way
         .json(json_file_path_multiline)
)

df_json_data_multiline.show(5)
```

```
parquet_file_loc="data/products.parquet"
```

```
df_json_data_multiline.write.parquet(parquet_file_loc)
```

```
text_file_path="/content/stocks.txt"
```

```
df = spark.read.csv(text_file_path, header=True, inferSchema=True)
```

```
df.show(5)
```

```
selected_column=df.select('id','name','price')
selected_column.show(5)
```

###Join Operation

```
df1=df.select('id','name','price')
```

```
df2=df.select('id','category')
```

```
joined_data=df1.join(df2,'id',"inner")
joined_data.show()
```

###Sorted

```
sorted_data=df.orderBy('price')
sorted_data.show()
```

###Handling Categorical Feature

```
distinct_data=df.select('category').distinct()
distinct_data.show()
```
